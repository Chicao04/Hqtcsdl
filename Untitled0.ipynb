```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
high = cv2.imread("highG.jpg", 0)
low = cv2.imread("lowG.jpg" , 0 )
```

```python
# hàm hstogram
def imhist(img):
  m,n = img.shape
  # hist chứa tần số của ảnh
  hist = [0]*256
  #với mọi giá trị của ảnh sẽ được thống kê vào danh sách hist
  for i in range(m):
    for j in range(n):
      hist[img[i,j]] += 1
  return np.array(hist)
```

```python
# Tạo vùng hiển thị cho ảnh kích thước 16:9
fig = plt.figure(figsize=(16,9))#tạo vùng vẽ 16:9
# Chia vùng hiển thị cho ảnh 2 hàng 2 cột
(ax1,ax2),(ax3,ax4) = fig.subplots(2,2)
ax1.imshow(high ,"gray")
ax1.set_title("High")
ax2.plot(imhist(high))
ax2.set_title("Histogram High")
ax3.imshow(low ,"gray")
ax3.set_title("Low")
ax4.plot(imhist(low))
ax4.set_title("Histogram Low")
```

```python
# cân bằng đa mức xám cho ảnh high và low
high_eq = cv2.equalizeHist(high)
low_eq = cv2.equalizeHist(low)
```

```python
fig = plt.figure(figsize=(16,9))
(ax1,ax2),(ax3,ax4) = fig.subplots(2,2)
ax1.imshow(high_eq,"gray")
ax1.set_title("High qua can bang")
ax2.plot(imhist(high_eq))
ax2.set_title("Histogram high")
ax3.imshow(low_eq,"gray")
ax3.set_title("High qua can bang")
ax4.plot(imhist(low_eq))
ax4.set_title("Histogram low")
```

```python
# phép closing
a4closing = cv2.morphologyEx(a4gray , cv2.MORPH_CLOSE , np.ones((5,5),np.uint8), iterations= 2)
plt.imshow(a4closing,'gray')
```

```python

kernel = np.ones((3,3),np.uint8)
closing = cv2.morphologyEx(a4can, cv2.MORPH_CLOSE, kernel)


plt.imshow(closing ,'gray')
```

```python
a4can = cv2.dilate(a4can ,cv2.getStructuringElement(cv2.MORPH_ELLIPSE , (3,3) ),iterations =1 )
plt.imshow(a4can, 'gray')
```

```python
a4Line  = cv2.HoughLinesP(a4can , 2, np.pi/180 ,300 , minLineLength= 100 , maxLineGap= 10)
lineA4 = []
a4copy = a4.copy()
# sau khi có các đường chúng ta tạm thời vẽ các đường để xem chúng ta có bao nhiêu đường
for line in a4Line:
  x1 , y1 , x2, y2 = line[0]
  lineA4.append(line[0])
  cv2.line(a4copy,(x1,y1),(x2,y2),(0,255,0 ), 2 )
  cv2.circle(a4copy,(x1,y1),5,(0, 0, 255) , -1)
  cv2.circle(a4copy,(x2,y2),5,(255,0,0), -1)
  cv2.putText(a4copy,f'{x1},{y1}',(x1,y1) , cv2.FONT_HERSHEY_SIMPLEX , 0.5 ,(0,0,255) , 1, cv2.LINE_AA)
  cv2.putText(a4copy,f'{x2},{y2}',(x2,y2) , cv2.FONT_HERSHEY_SIMPLEX , 0.5 ,(255,0,0) , 1, cv2.LINE_AA)
plt.imshow(a4copy)
print(len(lineA4))
for i in lineA4:
  print(i)
```

```python
lineA = [lineA4[0],lineA4[1],lineA4[2]]
lineA.append([366,55,469,342])
a4copy = a4.copy()
for i in lineA:
  x1,y1,x2,y2 = i
  cv2.line(a4copy ,(x1,y1) ,(x2,y2) ,(0,255,0) , 2)
  cv2.circle(a4copy,(x1,y1), 5 ,(0 ,0, 255) , -1)
  cv2.circle(a4copy,(x2,y2), 5,(255,0 ,0), -1)
  cv2.putText(a4copy,f'{x1} , {y1}',(x1,y1), cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255) ,1 ,cv2.LINE_AA)
  cv2.putText(a4copy,f'{x2}, {y2}' ,(x2, y2) ,cv2.FONT_HERSHEY_SIMPLEX ,0.5,(255,0,0) ,1,cv2.LINE_AA)
plt.imshow(a4copy)
```

```python
# suy ra phương trình đường thằng của các đường thẳng trên từ lineA4
# ax + by = c
def linear_ax(line):
  x1, y1, x2, y2 = line
  a = y2 - y1
  b = x1 - x2
  c = a*x1 + b*y1
  return a ,b, c
linearA4 = [linear_ax(i) for i in lineA]
print(linearA4)
```

```python
# công thức tiimf giao điểm của 2 đường thẳng
def intersection(line1,line2):
  a1 , b1, c1 = line1
  a2 , b2, c2 = line2
  d = a1*b2 - b1*a2
  x = (c1*b2 - c2*b1)/d
  y = (a1*c2 - a2*c1)/d
  return x,y
```

```python
a4copy = a4.copy()
# tìm giao điểm của các đường thẳng , loại bỏ các đường không cắt nhau
intersectionA4 = []
for i in range(len(linearA4)):
  for j in range(i+ 1, len(linearA4)):
    x,y = intersection(linearA4[i] ,linearA4[j])
    if x > 0 and x <a4.shape[1] and y > 0 and y < a4.shape[0]:
      intersectionA4.append((int(x) ,int(y)))

  print(intersectionA4)

  # ve cac giao diem tren anh va hien thi
  for point in intersectionA4:
    cv2.circle(a4copy,point,5, (0,0,255) , -1 )
    cv2.putText(a4copy,f'{point}' , point ,cv2.FONT_HERSHEY_SIMPLEX ,0.5 ,(255,0,0) ,1)

  plt.imshow(a4copy)
```

```python
a4copy =a4.copy()
cv2.line(a4copy, intersectionA4[0],intersectionA4[1],(0,255,0), 2)
cv2.line(a4copy, intersectionA4[0],intersectionA4[2],(0,255,0),2)
cv2.line(a4copy, intersectionA4[1],intersectionA4[3],(0,255,0), 2)
cv2.line(a4copy, intersectionA4[2] , intersectionA4[3] ,(0,255,0) , 2)
plt.imshow(a4copy)
```

```python
# thuc hien phep bien hinh de chuyen anh ve hiinh chu nhat
# ham tinh khoang cach
def distance(p1,p2):
  x1, y1 = p1
  x2, y2 = p2
  return np.sqrt((x1 - x2)**2 + (y1- y2)**2)
corA4 = intersectionA4
print(corA4)

 # tinh chieu dai va chieu rong cua intersectionA4
heightA4 = int(distance(corA4[0] ,corA4[2]))
widthA4 = int(distance(corA4[0], corA4[1]))
print(widthA4 ,heightA4)

  # Bien hinh theo tao do da cho
pts1 = np.float32(corA4)
pts2 = np.float32([[0,0] ,[0,widthA4],[heightA4,0] ,[heightA4,widthA4]])

a4w = cv2.getPerspectiveTransform(pts1,pts2)
a4warp =cv2.warpPerspective(a4,a4w,(heightA4 ,widthA4))
plt.imshow(a4warp)
```
